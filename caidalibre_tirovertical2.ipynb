```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,4.5,0.5)
print(t)
```

```python
x1 = 10*t+0.5*9.81*t**2
x2 = 0.5*9.81*t**2
print(x1)
print(x2)
plt.plot(t,x1,label='pelota')
plt.plot(t,x2,label='piedra')
plt.ylim(0, 50)
plt.xlabel("Tiempo")
plt.ylabel("altura")
plt.legend(loc='best')
plt.title('A')
plt.show()
```

```python
y1 = 50-10*t-0.5*9.81*t**2
y2 = 50-0.5*9.81*t**2
print(y1)
print(y2)
plt.plot(t,y1,label='pelota')
plt.plot(t,y2,label='piedra')
plt.ylim(0, 50)
plt.xlabel("Tiempo")
plt.ylabel("altura")
plt.legend(loc='best')
plt.title('B')
plt.show()
```

```python
t=np.arange(0,5,0.1)
z = 20*t-0.5*9.81*t**2
plt.plot(t,z,label='pelota')
plt.ylim(0, 25)
plt.xlabel("Tiempo")
plt.ylabel("altura")
plt.legend(loc='best')
plt.title('tiro vertical')
plt.show()
```

```python
v = 20-9.81*t
fig, ax = plt.subplots()
ax.plot(t,v,label='pelota',color='#ff7f0e')
ax.set_ylim(-30, 25)
plt.xlabel("Tiempo")
plt.ylabel("velocidad")
ax.axhline(y=0, color='black')
ax.axvline(x=0, color='black')
plt.legend(loc='best')
plt.title('tiro vertical')
plt.show()
```
