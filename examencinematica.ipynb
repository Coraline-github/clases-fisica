```python
import numpy as np
import matplotlib.pyplot as plt
```

# Examen 1

## MRU/MRUV
Un automóvil circula por una calle a velocidad constante, 30km/h por encima de la velocidad máxima permitida de 60km/h. En un cierto instante el auto pasa por delante de un patrullero en reposo que advierte la infracción y sale a perseguirlo con una aceleración constante de 5m/s2.
1. Determine cuánto se ha desplazado el automóvil en 4 segundos.
2. ¿Cuántos segundos tarda el patrullero en alcanzar la misma velocidad que el auto?¿En qué posición se encuentra?
3. ¿En qué instante alcanza el patrullero al auto?
4. Grafique la posición y la velocidad de ambos vehículos en función del tiempo



```python
# Constants 
velocity = 90  # km/h
velocity = velocity * 1000 / 3600  # Convert velocity from km/h to m/s
time = np.arange(0, 15, 0.1)  # Time range from 0 to 15 seconds

acceleration = 5  # m/s^2

# Calculate position car
position1 = velocity * time  # Convert velocity from km/h to km/s

# Calculate position police
position2 = 0.5 * acceleration * time**2  # Calculate position using the kinematic equation

# Plot position vs time
plt.plot(time, position1, label='Auto')
plt.plot(time, position2, label='Policia')
plt.xlabel('tiempo (s)')
plt.ylabel('Posición (m)')
plt.title('Posición del vehículo vs Tiempo')
plt.legend()
plt.grid(True)
plt.show()
```

```python
v = acceleration * time  # Calculate velocity using the kinematic equation

# Plot velocity vs time
plt.plot(time, velocity*np.ones(len(time)), label='Auto')
plt.plot(time, v, label='Policia')
plt.xlabel('tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.title('Velocidad del vehículo vs Tiempo')
plt.legend()
plt.grid(True)
plt.show()
```

```python
timeb = velocity / acceleration
posicionb = 0.5 * acceleration * timeb**2
timeb = np.round(timeb, 2)
posicionb = np.round(posicionb, 2)
tf = np.round(2 * velocity / acceleration, 2)

print('RESPUESTAS')
print('la velocidad del automovil es de', velocity, 'm/s')
print('a) la posicion del automovil a los 4 segundos es de',position1[40],'m')
print('b) el patrullero tarda', timeb, 'segundos en alcanzar la misma velocidad que el automovil')
print('el patrullero se encuentra en', posicionb, 'm')
print('c) el patrullero alcanza al automovil a los', tf, 'segundos')
```

## CAIDA LIBRE/ TIRO VERTICAL
Se lanza una pelota verticalmente hacia arriba, con una velocidad de 24m/s.
1. Calcular su posición y velocidad al cabo de 3s.
2. Determinar cuánto tarda en volver a pasar por el punto de partida.
3. Determinar la altura máxima y en cuantos segundos la alcanza.
4. Hallar en qué instante se encuentra a 15 m de altura.
5. Graficar la posición y la velocidad de la pelota en función del tiempo.


```python
# Constants
v0 = 24  # m/s
g = -9.8  # m/s^2

# Time range
time = np.arange(0, 5, 0.1)  # Time range from 0 to 5 seconds

# Calculate position
position = v0 * time + 0.5 * g * time**2

# Plot position vs time
plt.plot(time, position)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Position of Object vs Time')
plt.grid(True)
plt.show()
```

```python
v3 = np.round(v0 + g * 3, 2)
tf = np.round(-v0 / (0.5*g), 2)
t15 = np.round(np.roots([0.5*g, v0, -15]), 2)
hmax = np.round(np.max(position), 2)
tmax = np.round(-v0/g, 2)

print('RESPUESTAS')
print('a) la posicion del objeto a los 3 segundos es de', position[30], 'm')
print('y su velocidad es de', v3, 'm/s')
print('b) el objeto alcanza el suelo a los', tf, 'segundos')
print('c) la altura máxima alcanzada por el objeto es de', hmax, 'm', 'a los', tmax, 'segundos')
print('d) el objeto se encuentra a 15 m del suelo a los', t15[1], 's y a los', t15[0], 's')

print('e)')
# Plot velocity vs time
plt.plot(time, v0 + g*time)
plt.xlabel('tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.title('Velocidad vs Tiempo')
plt.grid(True)
plt.show()
```

## CAIDA LIBRE/ TIRO VERTICAL
Se lanza una piedra verticalmente hacia arriba desde un muelle a 280cm del agua, con una velocidad de 18m/s.
1. Calcular su posición y velocidad al cabo de 0.7s.
2. Determinar el instante en que entra en contacto con el agua.
3. Determinar la altura máxima y en cuantos segundos la alcanza.
4. Hallar en qué instante se hundió 5m (asumiendo que el agua no genera resistencia sobre la piedra).
5. Graficar la posición y la velocidad de la piedra en función del tiempo.

```python
# Constants
h0 = 280/100  # m
v0 = 18  # m/s
g = -9.8  # m/s^2

# Time range
time = np.arange(0, 5, 0.1)  # Time range from 0 to 5 seconds

# Calculate position
position = h0 + v0 * time + 0.5 * g * time**2

# Plot position vs time
plt.plot(time, position)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Position of Object vs Time')
plt.grid(True)
plt.show()
```

```python
v1 = v0 + g * 0.7
h1 = h0 + v0 * 0.7 + 0.5 * g * 0.7**2

tf = np.roots([0.5*g, v0, h0])

hmax = np.max(position)
tmax = -v0/g

thundido = np.roots([0.5*g, v0, h0+5])


print('RESPUESTAS')
print('a) la posicion del objeto a los 0.7 segundos es de', np.round(h1,2), 'm')
print('y su velocidad es de', np.round(v1,2), 'm/s')
print('b) el objeto alcanza el agua a los', np.round(tf[0], 2), 'segundos')
print('c) la altura máxima alcanzada por el objeto es de', np.round(hmax, 2), 'm', 'a los', np.round(tmax, 2), 'segundos')
print('d) el objeto se encuentra a -5 m de la superficie a los', np.round(thundido[0],2), 's')

print('e)')
# Plot velocity vs time
plt.plot(time, v0 + g*time)
plt.xlabel('tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.title('Velocidad vs Tiempo')
plt.grid(True)
plt.show()
```
